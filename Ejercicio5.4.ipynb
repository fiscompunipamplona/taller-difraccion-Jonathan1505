Nuestra capacidad para resolver detalles en observaciones astronómicas está limitada por la difracción de la luz
En nuestros telescopios. La luz de las estrellas puede ser tratada efectivamente como proveniente de una fuente puntual.
en el infinito. Cuando dicha luz, con una longitud de onda λ, atraviesa la abertura circular de un telescopio (que asumiremos que tiene un radio de unidad) y el telescopio lo enfoca en la zona focal.
plano, no produce un solo punto, sino un patrón de difracción circular que consiste en un punto central rodeado por una serie de anillos concéntricos. La intensidad de la luz en este patrón de difracción.
es dado por.
I(r) =((J1( kr ))/kr)**2
